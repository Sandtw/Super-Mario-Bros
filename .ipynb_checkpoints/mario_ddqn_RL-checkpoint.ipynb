Para hacer más eficiente el entrenamiento es importante hacerle algunas modificaciones a la información que el ambiente ofrece:

*   Cada acción tomada por el agente es repetida durante 4 frames consecutivos.
*   Se reduce cada frame de 240x256 a 84x84
*   El buffer junta 4 frames 
*   Los valores de los pixeles son escalados para estar en [0,1]
*   Reducimos el número de acciones de 256 a solo 5 usando RIGHT_ONLY


```python
class MaxAndSkipEnv(gym.Wrapper):
    def __init__(self, env=None, skip=4):
        """
        Devuelve un agregado de skip frames consecutivos
        y se ejecuta una misma acción en esos mismos
        skip frames.
        """
        super(MaxAndSkipEnv, self).__init__(env)

        self._obs_buffer = collections.deque(maxlen=skip)
        self._skip = skip

    # Paso al que se le pasa una accion y se aplica  a los 4 frames, y regresa 4 observaciones, y las observaciones
    # se van a guardar en un buffer, y de los 4 pasos de tiempo se acumula las recompensas
    # y luego va hacer el maximo de esos 4 frames apilados.
    # Regresa el max_frame, y el total de recompensa
    def step(self, action):
        total_reward = 0.0
        done = None
        for _ in range(self._skip):
            obs, reward, done, info = self.env.step(action)
            self._obs_buffer.append(obs)
            total_reward += reward
            if done:
                break
        max_frame = np.max(np.stack(self._obs_buffer), axis=0)
        return max_frame, total_reward, done, info

    def reset(self):
        """Limpia el buffer y regresa la primer observación"""
        self._obs_buffer.clear()
        obs = self.env.reset()
        self._obs_buffer.append(obs) # Esa obs se va añadir al buffer
        return obs
    

#
class PreprocessFrame(gym.ObservationWrapper):
    """
    Toma un frame de 240x256x3 y lo convierte a 84x84x1.
    """
    def __init__(self, env=None):
        super(PreprocessFrame, self).__init__(env)
        self.observation_space = gym.spaces.Box(low=0, high=255, shape=(84, 84, 1), dtype=np.uint8)

    def observation(self, obs):
        return PreprocessFrame.process(obs)

    @staticmethod
    def process(frame):
        #Reto 2
        #Realiza el preprocesamiento del frame:
        #   Reducción a un color, resize a 110x84 y recorte a 84x84.
        #   Finalmente expande el array agregando una dimensión al final
        #   y convierte todo a tipo de dato np.unit8 (enteros).

        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        frame = cv2.resize(frame, (84, 110), interpolation = cv2.INTER_AREA)
        frame = frame[18:102,:] 
        frame = np.reshape(frame, [84, 84, 1])
        frame.astype(np.uint8) # pixeles enteros
        return frame


class ChannelFirst(gym.ObservationWrapper):
    """
    Cambia el canal a la primera dimensión.
    Se tiene que hacer un tensor con 4 pasos de tiempos para que de la sensacion de movimiento
    combiene trabajarlo cuando el canal es 1ro
    """
    def __init__(self, env):
        super(ChannelFirst, self).__init__(env)
        old_shape = self.observation_space.shape
        self.observation_space = gym.spaces.Box(low=0.0,
                                                high=1.0,
                                                shape=(old_shape[2], old_shape[0], old_shape[1]), dtype=np.float32)

    def observation(self, observation):
        return np.moveaxis(observation, 2, 0)
        #observation: [C,H,W]

# Pongo las nuevas observaciones en el buffer, y las anteriores ultimas se sacan
class BufferWrapper(gym.ObservationWrapper):
    def __init__(self, env, n_steps):
        super(BufferWrapper, self).__init__(env)
        old_space = env.observation_space
        self.observation_space = gym.spaces.Box(old_space.low.repeat(n_steps, axis=0),
                                                old_space.high.repeat(n_steps, axis=0), dtype=np.float32)
    def reset(self):
        self.buffer = np.zeros_like(self.observation_space.low, dtype=np.float32)
        return self.observation(self.env.reset())

    def observation(self, observation):
        self.buffer[:-1] = self.buffer[1:]
        self.buffer[-1] = observation
        return self.buffer

    
class ScaledFloatFrame(gym.ObservationWrapper):
    """Escalar pixeles en el rango [0,1]"""
    def observation(self, obs):
        return np.array(obs).astype(np.float32) / 255.0


class ChannelLast(gym.ObservationWrapper):
    """
    Cambia a la última dimensión el canal.
    """
    def __init__(self, env):
        super(ChannelLast, self).__init__(env)
        old_shape = self.observation_space.shape
        self.observation_space = gym.spaces.Box(low=0.0,
                                                high=1.0,
                                                shape=(old_shape[1], old_shape[2], old_shape[0]), dtype=np.float32)

    def observation(self, observation):
        #Permuta la dimensiones 0 y 2 para tener una observación con shape [H, W, C]
        return np.moveaxis(observation, 0, 2)
        #observation: [H,W,C]

# Pasamos el ambiente de gym y se computa lo que esta adentro
def make_env(env): 
    env = MaxAndSkipEnv(env)
    env = PreprocessFrame(env)
    env = ChannelFirst(env)
    env = BufferWrapper(env, 4)
    env = ScaledFloatFrame(env)
    env = ChannelLast(env)
    return JoypadSpace(env, RIGHT_ONLY)
```

```python
#Crea el ambiente 'SuperMarioBros-1-1-v0', pásalo como parámetro
#   en la función make_env() e imprime el espacio de estados y acciones.
env = gym_super_mario_bros.make('SuperMarioBros-1-1-v0')
env = make_env(env)

# Espacio de estados
print(env.observation_space.shape)

# Espacio de acciones
print(env.action_space.n)
```

Tip:
* Cada pixel toma valores entre 0 y 255
* Cada estado es una secuencia de 4 imágenes consecutivas de 84x84
* 5 posibles acciones para cada estado

Sol: 256^(84x84x4) * 5

## Agente

```python
# Arquitectura Neuronal
def build_network(input_shape, output_shape):
    model = models.Sequential()
    # Una CNN de 3 capas convolucionales.
    # Capa 1: 32 filtros de 8x8 con pasos de 4x4, sin padding, ReLU y entrada input_shape.
    # Capa 2: 64 filtros de 4x4 con pasos de 2x2, sin padding, ReLU.
    # Capa 3: 64 filtros de 3x3 con pasos de 1x1, sin padding, ReLU.
    #Finalmente haz flatten, crea una capa oculta de tamaño 512 con activación ReLU
    # y una capa de salida sin activación de tamaño output_shape.
    model.add(layers.Conv2D(filters = 32,
                            kernel_size = (8,8),
                            strides = (4,4),
                            padding = 'valid',
                            activation = 'relu',
                            input_shape = input_shape))

    model.add(layers.Conv2D(
              filters = 64,
              kernel_size = (4,4),
              strides = (2,2),
              padding = 'valid',
              activation = 'relu'           
             ))

    model.add(layers.Conv2D(
              filters = 64,
              kernel_size = (3,3),
              strides = (2,2),
              padding = 'valid',
              activation = 'relu'))

    model.add(layers.Flatten())
    model.add(layers.Dense(512, activation='relu'))
    model.add(layers.Dense(output_shape, activation = None))
    # Activacion lineal pq quiero que me regresen valores arbitrarios
    # output_shape, nro de clases
    return model
```

```python
#   Crea la clase ReplayBuffer con los métodos add_exp
#   para añadir experiencias al buffer y sample_exp
#   para muestrar experiencias del buffer.

class ReplayBuffer():
    def __init__(self, buffer_size, observation_space):
        self.buffer_size = buffer_size
        self.pointer = 0
        
        # Crear un numpy attay de ceros del tamaño y tipo adecuado para almacenar
        # cada uno de los siguientes elementos:
        self.state_memory = np.zeros((self.buffer_size, *observation_space), dtype = np.float32)
        self.action_memory = np.zeros((self.buffer_size), dtype = np.int32)
        self.reward_memory = np.zeros((self.buffer_size), dtype = np.float32)
        self.next_state_memory = np.zeros((self.buffer_size, *observation_space), dtype = np.float32)
        self.done_memory = np.zeros((self.buffer_size), dtype = np.bool)
        
    def add_exp(self, state, action, reward, next_state, done):
        idx = self.pointer % self.buffer_size
        
        # Haz que cada elemento (state, action, etc.) se almacene en el indice
        # y buffer adecuado. Al terminar este proceso, el puntero 'pointer'
        # debe incrementarse en una unidad
        self.state_memory[idx] = state
        self.action_memory[idx] = action
        self.reward_memory[idx] = reward
        self.next_state_memory[idx] = next_state
        self.done_memory[idx] = 1 - int(done)
        self.pointer += 1

    def sample_exp(self, batch_size):
        max_mem = min(self.pointer, self.buffer_size)
        
        # Crea el codigo necesario para obtener una lista aleatoria de indices,
        # los indices deberam ser enteros entre 0 y max_mem y la muestra debera
        # ser sin reemplazo y de tamaño batch_size.
        # Ej: Si max_mem = 19 y batch_size = 2, entonces un batch puede ser 
        # el array [6,2}, donde 6 y 2 son los dos numeros aleatorios escogidos
        # entre 0 y 10.
        # Tip: puedes usar np.random.
        
        batch = np.random.choice(max_mem, batch_size, replace = False)
        
        states = self.state_memory[batch]
        actions = self.action_memory[batch]
        rewards = self.reward_memory[batch]
        next_states = self.next_state_memory[batch]
        dones = self.done_memory[batch]        
        
        return states, actions, rewards, next_states, dones
 
```

```python
class DDQNAgent():
    def __init__(self, lr, gamma, action_space, observation_space,
                 epsilon, epsilon_decay, epsilon_min, batch_size,
                 buffer_size, replace_target_net):
    
        self.gamma = gamma
        self.action_space = action_space
        self.epsilon = epsilon
        self.epsilon_decay = epsilon_decay
        self.epsilon_min = epsilon_min
        self.batch_size = batch_size
        self.replace_target_net = replace_target_net
        
    
        #   Usa build_network para crear ambas NN (online y target),
        #   selecciona correctamente el input_shape y output_shape
        #   de cada una. Usa el 'compile' en la NN correspondiente,
        #   elige el optimizador y loss adecuado. Usa lr como parámetro
        #   para learning_rate. Finalmente, crea un buffer usando
        #   ReplayBuffer.
        self.Q_net = build_network(input_shape = observation_space, #84x84x4
                                  output_shape = action_space) # S
        
        self.T_net = build_network(input_shape = observation_space, 
                                  output_shape = action_space)
        
        
        self.Q_net.compile(optimizer=Adam(learning_rate=lr), loss='huber_loss')

        self.memory = ReplayBuffer(buffer_size, observation_space)
        

    def save_experience(self, state, action, reward, next_state, done):
        self.memory.add_exp(state, action, reward, next_state, done)
         

    def choose_action(self, state, traning_mode):

        #   Crea una política épsilon greedy para escoger acciones.
        #   En testing, el agente debe elegir un acción al azar con
        #   probabilidad igual o menor a epsilon_min y una acción greedy
        #   de otra forma. En training, el agente debe elegir una acción
        #   al azar con probabilidad igual o menor a épsilon y greedy
        #   de otra forma. 

        #En testing:
        if not traning_mode:
            logits = self.Q_net.predict(np.array([state]))
            action = np.argmax(logits)
            return action

        #En entrenamiento:
        if traning_mode:
            if np.random.rand() <= self.epsilon:
                return np.random.choice([i for i in range(self.action_space)])
            else:
                logits = self.Q_net.predict(np.array([state]))
                action = np.argmax(logits)
                return action
 
      

    def learn(self):
        if self.memory.pointer < self.batch_size:
            #El aprendizaje comienza hasta haber suficientes
            #   experiencias en el buffer
            return 
        
        if self.memory.pointer % self.replace_target_net == 0:
            #Actualizar T_net con parámetros de Q_net
            self.T_net.set_weights(self.Q_net.get_weights())
        
        #Toma un batch de experiencias del buffer
        states, actions, rewards, next_states, dones = \
            self.memory.sample_exp(self.batch_size)
        
        #DDQN:
        #w <- w + \alpha * gradiente (r + \gamma T_net(s', a*) - Q_net(s,a))^2
        #a* = argmax Q_net(s',a')

        #Se requieren 3 cosas:
        #   * predicción actual = Q_net(s,a)
        #   * a* = argmax Q_net(s',a')
        #   * target = r + \gamma T_net(s', a*)

        #Predicción actual = Q_qnet(s,a)
        pred = self.Q_net.predict(states)

        #a* = argmax Q_net(s',a')
        a_max = np.argmax(self.Q_net.predict(next_states), axis=1)
        
        #Estimated TD-target = r + \gamma T_net(s', a*)
        batch_index = np.arange(self.batch_size, dtype=np.int32)
        target = np.copy(pred)
        
        target[batch_index, actions] = rewards + \
            self.gamma * self.T_net.predict(next_states)[batch_index, a_max] * dones

        
        self.Q_net.train_on_batch(states, target)
        
        self.epsilon = self.epsilon * self.epsilon_decay if self.epsilon > \
            self.epsilon_min else self.epsilon_min
        

    def load_model(self):
        print('Loading pretrained model ...')
        self.Q_qnet = load_model("model.h5")
        self.Q_tnet = load_model("model.h5")
        print(' ... Done.')
```

```python
def run(training_mode, num_episodios):
    env = gym_super_mario_bros.make('SuperMarioBros-1-1-v0')
    env = make_env(env)
    
    #Pon los valores que consiederes prudentes para cada
    #   uno de los hiperparámetros.
    agent = DDQNAgent(lr = 0.00025,
                     gamma = 0.990,
                     action_space = env.action_space.n,
                     observation_space = env.observation_space.shape,
                     epsilon = 1.0,
                     epsilon_decay = 0.992,
                     epsilon_min = 0.005,
                     batch_size = 32,
                     buffer_size = 10000, #3000
                     replace_target_net = 5000) #5000
    
    if not training_mode:
        agent.load_model()

    total_rewards = []
    for ep_num in tqdm(range(num_episodios)):
        state = env.reset()
        total_reward = 0
        
        while True:
            if not training_mode:
                env.render()
                time.sleep(0.02)
            
            action = agent.choose_action(state, training_mode)
            
            next_state, reward, done, info = env.step(action)
            total_reward += reward
            
            if training_mode:
                agent.memory.add_exp(state, action, reward, next_state, done)
                agent.learn()
            state = next_state
            if done:
                break
        
        total_rewards.append(total_reward)

        if (ep_num + 1) % 50 == 0:
            moving_reward = np.mean(total_rewards[max(0,ep_num-50):(ep_num+1)])
            print(f'Episode: {ep_num + 1}, Total reward: {total_rewards[-1]}, Moving reward: {moving_reward}')
     
    
    if training_mode:
        agent.Q_net.save("model_i.h5")
        agent.T_net.save("target_model_i.h5")

        plt.title("Recompensa total por episodio")
        plt.plot(total_rewards[:50-1] +
                 (np.convolve(total_rewards, np.ones(50), 'Valid')/50).tolist())
        plt.savefig('mario_ddqn_i.png')
    
    env.close()
```

```python
run(training_mode=True, num_episodios=100)
```
